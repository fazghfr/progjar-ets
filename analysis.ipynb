### Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# importing a csv file to pandas dataframe
df = pd.read_csv('data.csv')
```

```python
# creating 4 groups of data
df_thread_secure = df[df['type'] == 'TSCR']
df_thread_not_secure = df[df['type'] == 'TNSCR']

df_process_secure = df[df['type'] == 'PSCR']
df_process_not_secure = df[df['type'] == 'PNSCR']
```

```python
# seeing the data summary
# print(df_thread_secure.describe())
print(df_process_not_secure.describe())
```

### Compare Secure vs non-secure Processes

```python

# Get the elapsed time from the secure process dataframe
df_process_secure_temp = df_process_secure[['elapsed_time', 'jumlah concurrency']]

# Get the elapsed time from the non-secure process dataframe
df_process_not_secure_temp = df_process_not_secure[['elapsed_time', 'jumlah concurrency']]

# Plotting the data
plt.bar(df_process_secure_temp['jumlah concurrency'], df_process_secure_temp['elapsed_time'], width=15, label='Secure Process')
plt.bar(df_process_not_secure_temp['jumlah concurrency'] + 15, df_process_not_secure_temp['elapsed_time'], width=15, label='Non-Secure Process')

# Adding labels and title
plt.xlabel('Jumlah Concurrency')
plt.ylabel('Elapsed Time')
plt.title('Comparison of Elapsed Time between Secure and Non-Secure Processes')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python

# Get the elapsed time from the secure process dataframe
df_process_secure_temp = df_process_secure[['failed_transaction', 'jumlah concurrency']]

# Get the elapsed time from the non-secure process dataframe
df_process_not_secure_temp = df_process_not_secure[['failed_transaction', 'jumlah concurrency']]

# Plotting the data
plt.bar(df_process_secure_temp['jumlah concurrency'], df_process_secure_temp['failed_transaction'], width=15, label='Secure Process')
plt.bar(df_process_not_secure_temp['jumlah concurrency'] + 15, df_process_not_secure_temp['failed_transaction'], width=15, label='Non-Secure Process')

# Adding labels and title
plt.xlabel('Jumlah Concurrency')
plt.ylabel('failed_transaction')
plt.title('Comparison of Failed Request between Secure and Non-Secure Processes')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python

# Get the elapsed time from the secure process dataframe
df_process_secure_temp = df_process_secure[['response_time', 'jumlah concurrency']]

# Get the elapsed time from the non-secure process dataframe
df_process_not_secure_temp = df_process_not_secure[['response_time', 'jumlah concurrency']]

# Plotting the data
plt.bar(df_process_secure_temp['jumlah concurrency'], df_process_secure_temp['response_time'], width=15, label='Secure Process')
plt.bar(df_process_not_secure_temp['jumlah concurrency'] + 15, df_process_not_secure_temp['response_time'], width=15, label='Non-Secure Process')

# Adding labels and title
plt.xlabel('Jumlah Concurrency')
plt.ylabel('response_time')
plt.title('Comparison of response_time between Secure and Non-Secure Processes')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```
